```python
# 영상 평균 term - 일년치 term을 평균내서 사용 ⇒ 성실도
# 영상 갯수 → 월 영상 갯수 ⇒ 성실도
total_year = pd.read_csv("Untitled_Folder/total_year_video_df.csv")
total_year.head()
```

```python
total_year['year'] = total_year['업데이트 날짜'].str[:4]
total_year['month'] = total_year['업데이트 날짜'].str[5:7]
total_year['day'] = total_year['업데이트 날짜'].str[8:]
total_year.head()
```

## 유튜버 월 별 영상개수 

```python
monthcount_df = total_year.groupby(['채널명','year','month']).count().reset_index()[['채널명','year','month','Unnamed: 0']]
monthcount_df.columns
```

```python
monthcount_df.columns = ['채널명', 'year', 'month', 'count']
monthcount_df.head()
```

```python
monthcount_df.info()
```

```python
monthcount_df['날짜'] = monthcount_df['year'].astype(str) + '/' + monthcount_df['month'].astype(str)
monthcount_df
```

```python
mt = pd.read_excel('mater_table.xlsx', index_col=0)
mt
```

```python
monthcount_df = monthcount_df.drop(['year','month'],axis = 1)
monthcount_df.columns
```

```python
monthcount_df.columns = ['유투버', '영상개수', '월']
```

```python
mon_count = pd.merge(mt,monthcount_df,how = 'outer')
```

```python
mon_count = mon_count.fillna(0)
mon_count.info()
```

```python
mon_count.to_csv('mon_count.csv')
```
