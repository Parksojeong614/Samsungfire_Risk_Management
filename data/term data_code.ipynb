```python
import pandas as pd
```

```python
total_video = pd.read_csv("Untitled_Folder/total_video_df.csv")
total_video.head()
```

```python
total_video['업데이트 날짜'] = pd.to_datetime(total_video['업데이트 날짜'])
total_video.info()
```

```python
name = total_video['채널명'].unique()
```

```python
# 빈 데이터 프레임 생성
df = pd.DataFrame(columns=["name", "날짜", "평균"])
```

```python
for j, k in enumerate(name):
    date_df = total_video[total_video['채널명'] == name[j]][['업데이트 날짜','채널명']].sort_values(by='업데이트 날짜', ascending=True)
    new_date = date_df['업데이트 날짜'][1:].reset_index()[['업데이트 날짜']]
    new_date.loc[new_date.index.max() + 1] = new_date.iloc[-1]
    new_date.columns = ['new']
    new_date2 = pd.concat([date_df.reset_index(drop = True),new_date],axis = 1)
    new_date2['term'] = new_date2['new'] - new_date2['업데이트 날짜']
    
        # 시작 날짜와 종료 날짜 리스트 생성 (datetime 형식)
    start_date = [pd.to_datetime('2021-09-01'), pd.to_datetime('2021-10-01'), pd.to_datetime('2021-11-01'), pd.to_datetime('2021-12-01'), pd.to_datetime('2022-01-01'), pd.to_datetime('2022-02-01'), pd.to_datetime('2022-03-01'), pd.to_datetime('2022-04-01'),
                  pd.to_datetime('2022-05-01'), pd.to_datetime('2022-06-01'), pd.to_datetime('2022-07-01'), pd.to_datetime('2022-08-01'), pd.to_datetime('2022-09-01')]
    end_date = [pd.to_datetime('2022-08-31'), pd.to_datetime('2022-09-30'), pd.to_datetime('2022-10-31'), pd.to_datetime('2022-11-30'), pd.to_datetime('2022-12-31'), pd.to_datetime('2023-01-31'), pd.to_datetime('2023-02-28'), pd.to_datetime('2023-03-31'),
                pd.to_datetime('2023-04-30'), pd.to_datetime('2023-05-31'), pd.to_datetime('2023-06-30'), pd.to_datetime('2023-07-31'), pd.to_datetime('2023-08-31')]
    for i in range(len(start_date)):
        filtered_df = new_date2[(new_date2['업데이트 날짜'] >= start_date[i]) & (new_date2['업데이트 날짜'] <= end_date[i])]
        average_tem = filtered_df['term'].mean()
        new_data = {
            'name': name[j],  # 실제 이름으로 대체
            '날짜': end_date[i],
            '평균': average_tem 
        }
        # 새로운 데이터를 데이터 프레임에 추가
        df = df.append(new_data, ignore_index=True)
        
```

```python
total_year = pd.read_csv("Untitled_Folder/total_year_video_df.csv")
total_year['채널명']
```

```python
# 'name' 변수의 값이 total_year_channels에 있는 것만 추출
filtered_df = df[df['name'].isin(total_year['채널명'].unique())]
filtered_df
```

```python
filtered_df.info()
```

```python
# '날짜' 열을 datetime64 형식으로 변환
filtered_df['날짜'] = pd.to_datetime(filtered_df['날짜'])

# '날짜' 열을 '년도/월' 형식으로 포맷팅하여 다시 'filtered_df'에 저장
final_df = filtered_df.copy()
final_df['날짜'] = final_df['날짜'].dt.strftime('%Y/%m')
final_df
```

```python
# '평균' 열을 timedelta 형식으로 변환
final_df['평균'] = pd.to_timedelta(final_df['평균'])

final_df['일'] = final_df['평균'].dt.days

final_df
```

```python
final_df = final_df.drop(['평균'],axis = 1)
final_df.columns
```

```python
final_df.columns = ['유튜버', '월', 'term']
final_df
```
